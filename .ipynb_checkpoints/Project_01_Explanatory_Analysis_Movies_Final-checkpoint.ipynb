# The best and the worst movies... (Part 1)

```python
# Needed libraries
from IPython.display import HTML
pd.options.display.max_colwidth = 200
```

```python
# Making the copy of dataset
movies_best = movies[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
movies_best
```

```python
# Getting the profit details
movies_best["profit_musd"] = movies.revenue_musd.sub(movies.budget_musd)
movies_best["return"] = movies.revenue_musd.div(movies.budget_musd)
```

```python
movies_best
```

```python
# Best movies details
movies_best.columns = ["", "Title", "Budget", "Revenue", "Votes", 
                   "Average Rating", "Popularity", "Profit", "ROI"]
```

```python
movies_best.set_index("Title", inplace = True)
```

```python
movies_best
```

```python
movies_best.iloc[0,0]
```

```python
# Preparing for subset
subset = movies_best.iloc[:5, :2]
subset
```

```python
# Lets see though HTMl
HTML(subset.to_html(escape=False))
```

```python
# Sorting it by best values
movies_best.sort_values(by = "Average Rating", ascending = False)
```

```python
# Sorting by "ROI"
movies_best.sort_values(by = "ROI", ascending = False)
```

```python
# Locating the best out of the abouve details
movies_best.loc[movies_best.Budget >= 5].sort_values(by = "ROI", ascending = False)
```

```python
# Filling null values
movies_best.Budget.fillna(0, inplace = True)
movies_best.Votes.fillna(0, inplace = True)
```

```python
movies_best.info()
```

```python
def best_worst(n, by, ascending = False, min_bud = 0, min_votes = 0):
    
    
    movies2 = movies_best.loc[(movies_best.Budget >= min_bud) & (movies_best.Votes >= min_votes), 
                      ["", by]].sort_values(by = by, ascending = ascending).head(n).copy()
    
    return HTML(movies2.to_html(escape=False))
```

# What are the most common Words in Movie Titles and Taglines?

```python
# Dropping the few columns
title = movies.title.dropna()
overview = movies.overview.dropna()
tagline = movies.tagline.dropna()
```

```python
title
```

```python
' '.join(title)
```

```python
title_corpus = ' '.join(title)
overview_corpus = ' '.join(overview)
tagline_corpus = ' '.join(tagline)
```

```python
tagline_corpus
```

```python
title_wordcloud = WordCloud(background_color='white', height=2000, width=4000, max_words= 200).generate(title_corpus)
title_wordcloud
```

```python
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud, interpolation= "bilinear")
plt.axis('off')
plt.show()
```

```python
# For tagline
tagline_wordcloud = WordCloud(background_color='white', height=2000, width=4000).generate(tagline_corpus)
plt.figure(figsize=(16,8))
plt.imshow(tagline_wordcloud, interpolation= "bilinear")
plt.axis('off')
plt.show()
```

```python
# For overview
overview_wordcloud = WordCloud(background_color='white', height=2000, width=4000).generate(overview_corpus)
plt.figure(figsize=(16,8))
plt.imshow(overview_wordcloud, interpolation= "bilinear")
plt.axis('off')
plt.show()
```

```python
movies.info()
```

#### From above its difficult to get details, so lets split it and make it more understandable.

```python
movies.cast.str.split("|", expand = True)
```

```python
act = movies.cast.str.split("|", expand = True)
act
```

```python
# Only getting id along with cast details
act.stack().reset_index(level=1, drop=True).to_frame()
```

```python
act = act.stack().reset_index(level=1, drop=True).to_frame()
```

```python
act
```

```python
act.columns = ["Actor"]
```

```python
act = act.merge(movies[["title", "revenue_musd", "vote_average", "popularity"]],
                how = "left", left_index = True, right_index = True)
```

```python
act
```

#### So from here we have got id, actors details along with vote & popularity details

# Most Successful Actors (Part 2)

```python
act
```

```python
act.Actor.nunique()
```

```python
act.Actor.value_counts().head(20)
```

```python
# Checking for actor who has palyed most of the movies
plt.figure(figsize = (12, 8))
act.Actor.value_counts().head(20).plot(kind='bar', fontsize = 15)
plt.title("Most Active Actors",fontsize = 20)
plt.ylabel("Number of Movies", fontsize = 15)
plt.show()
```

```python
agg = act.groupby("Actor").agg(Total_Revenue = ("revenue_musd", "sum"), 
                               Mean_Revenue = ("revenue_musd", "mean"),
                               Mean_Rating = ("vote_average", "mean"), 
                               Mean_Pop = ("popularity", "mean"), 
                               Total_Movies = ("Actor", "count"))
```

```python
agg.nlargest(10, "Total_Movies")
```

```python
agg.nlargest(10, "Total_Revenue")
```

```python
plt.figure(figsize = (12, 8))
agg.Total_Revenue.nlargest(10).plot(kind='bar', fontsize = 15)
plt.title("Total Revenue",fontsize = 20)
plt.ylabel("Revenue (in MUSD)", fontsize = 15)
plt.show()
```

```python
agg.Mean_Revenue.nlargest(10)
```
